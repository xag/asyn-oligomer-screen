```python
#@title ▶️  Donate computing power to Parkinson's research &mdash; press play, then click the link it shows. No setup, no sign-up.
# One cell does it all: installs, fetches the project, links your session via
# the website (so no email is typed here), then runs short simulations and
# sends the results back. Use a GPU runtime: Runtime > Change runtime type > GPU.
import subprocess, sys, os

HEALTH_URL = 'https://health-two-iota.vercel.app/screen'  # the site; nothing to edit
TASKS = 5                                                  # how many simulations this session

print('Setting up (~2 min)…', flush=True)
subprocess.run([sys.executable, '-m', 'pip', '-q', 'install', 'openmm', 'requests',
                'git+https://github.com/openmm/pdbfixer.git'], check=True)
if not os.path.isdir('asyn-oligomer-screen'):
    subprocess.run(['git', 'clone', '-q', 'https://github.com/xag/asyn-oligomer-screen'], check=True)
sys.path.insert(0, 'asyn-oligomer-screen/screen')

import contribute_client as c
token, _ = c.pair(HEALTH_URL)
done = 0
for _ in range(TASKS):
    msg = c.run_once(HEALTH_URL, token)
    print(msg, flush=True)
    if not msg.startswith('submitted'):
        break
    done += 1
print(f'\nThank you — ran {done} task(s).', flush=True)
```
