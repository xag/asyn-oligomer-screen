# Contribute GPU time to the α-synuclein screen

Run short physics simulations that test whether a molecule loosens or holds the toxic α-synuclein shapes linked to Parkinson's. Each run picks up whatever the screen needs next and sends the result back. Stop any time — an unfinished run is simply reassigned.

1. **Runtime → Change runtime type → GPU.**
2. Run **Set up** once.
3. Get your token (next cell) — or skip to run anonymously.
4. *Optional:* pick a molecule to target — list the candidates, then name it in **Contribute**.
5. Run **Contribute**.

**Your token is your trust.** It ties each run to you and builds a trust score — results from trusted contributors carry more weight in the screen's consensus, so your compute counts for more. No token? Runs are anonymous: they still count, at the lowest trust.

Getting one stays inside this notebook: the **Get a token** cell emails you a link; click it, come back, and the cell has already captured your token (nothing to copy). Already have one from a past visit? Paste it into **Contribute** — tokens last 90 days.

**Picking a molecule is optional.** Leave it blank and you get whatever the screen needs next — the simplest way to help. Choose one only if you want your compute to go toward a specific candidate; run **List candidate molecules** to see what has work right now.

```python
#@title Get a token — credits your runs (optional) { display-mode: "form" }
#@markdown Enter your email and run. We email you a link; click it, then return here — your token is captured automatically, no copy-paste. Skip to run anonymously.
EMAIL = ""  #@param {type:"string"}
import os, time, hashlib, secrets, requests
from contribute_client import DEFAULT_BASE_URL
email = EMAIL.strip()
if not email:
    print("No email entered — skip this to run anonymously, or add your email above.")
else:
    claim = secrets.token_urlsafe(18)
    claim_hash = hashlib.sha256(claim.encode()).hexdigest()
    r = requests.post(f"{DEFAULT_BASE_URL}/api/screen/v1/token",
                      data={"email": email, "claim_hash": claim_hash}, timeout=30)
    vid = (r.json() if r.ok else {}).get("id")
    if not vid:
        print(f"Couldn't start (HTTP {r.status_code}) — try again shortly.")
    else:
        print("Sent. Click the link in the email from health, then come back here — waiting", end="", flush=True)
        token = None
        for _ in range(120):  # ~10 min
            time.sleep(5)
            p = requests.get(f"{DEFAULT_BASE_URL}/api/screen/v1/token-poll",
                             params={"id": vid, "claim": claim}, timeout=30)
            if (p.json() if p.ok else {}).get("status") == "ready":
                token = p.json()["token"]
                break
            print(".", end="", flush=True)
        if token:
            os.environ["ASYN_CONTRIB_TOKEN"] = token
            print("\nVerified — your token is set for this session. Run Contribute below.")
        else:
            print("\nNo click yet. Re-run this cell to try again (the link lasts 30 min).")
```
